## Análise de Componentes Principais em R - Introdução

```r
install.packages("corrplot")

library(corrplot)

R <- read.table("http://conteudo.icmc.usp.br/pessoas/cibele/Dados/Substancias.txt")

colnames(R) <- rownames(R) <- c("cigarro", "cerveja", "vinho", "licor", "cocaina", "tranquilizantes", "medicamentos", "heroina", "maconha", "haxixe", "inalantes",
                 "alucinogenos", "anfetaminas")
```

```r

round(R,2);
```

```r
corrplot(as.matrix(R))
```

```r
ACP <- princomp(R, cor=TRUE)
```

```r

summary(ACP,2)
```

```r
eigen(R)$values

eigen(R)$values/sum(eigen(R)$values)

eigen(R)$vectors

round(eigen(R)$vectors,2)[,1]
```

```r

summary(prcomp(R))
```

```r

# --- Análise da primeira componente principal --- #
# --- Quais os maiores pesos? --- #
# --- Que nome você daria à primeira componente principal? --- #

t(t(round(prcomp(R)[[2]][,1],2)))
```

```r
names(prcomp(R)[[2]][,1])
```

```r
# Ordenação das componentes em PCA1

ordem = order(round(prcomp(R)[[2]][,1],2),decreasing = T)

names(prcomp(R)[[2]][,1])[ordem]
```

```r
# Quantas componentes escolher?

# --- Scree plot --- #

plot(eigen(R)[[1]], type="b", pch=16, main="Scree plot", ylab="autovalor", xlab="ordem da componente")
```
